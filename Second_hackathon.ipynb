```python
def count_outliers_iqr(df):
    outlier_counts = {}
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - 1.5 * IQR
            upper = Q3 + 1.5 * IQR
            outliers = df[(df[col] < lower) | (df[col] > upper)]
            outlier_counts[col] = outliers.shape[0]
    return pd.DataFrame(list(outlier_counts.items()), columns=["Feature", "Outlier Count"])

count_outliers_iqr(x)
```

```python
def cap_outliers_specific(df, columns):
    df_capped = df.copy()
    for col in columns:
        if pd.api.types.is_numeric_dtype(df_capped[col]):
            lower = df_capped[col].quantile(0.01)
            upper = df_capped[col].quantile(0.99)
            df_capped[col] = df_capped[col].clip(lower, upper)
    return df_capped
```

```python
columns_to_cap = ['BMXBMI', 'LBXGLU', 'LBXGLT','LBXIN']
x = cap_outliers_specific(x, columns_to_cap)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.preprocessing import RobustScaler
nhanes_columns = ['BMXBMI','LBXGLU','LBXGLT','LBXIN']
scaler = RobustScaler()
x_train[nhanes_columns] = scaler.fit_transform(x_train[nhanes_columns])
```

```python
x_test[nhanes_columns] = scaler.transform(x_test[nhanes_columns])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
def model_evaluation(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    print('For training dataset: ')
    y_pred = model.predict(x_train)
    print('-' * 65)
    print(confusion_matrix(y_pred,y_train))
    print('-' * 65)
    print(classification_report(y_pred,y_train))
    print('-' * 65)
    print('f1-score:',f1_score(y_pred,y_train))
    print('-' * 65)
    print('For testing dataset: ')
    y_pred_test = model.predict(x_test)
    print('-' * 65)
    print(confusion_matrix(y_pred_test,y_test))
    print('-' * 65)
    print(classification_report(y_pred_test,y_test))
    print('-' * 65)
    print('f1-score:', f1_score(y_pred_test,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight = 'balanced', max_iter = 1000)
model_evaluation(lr,x_train,x_test,y_train,y_test)
```

```python
from sklearn.svm import SVC
support_vector = SVC()
model_evaluation(support_vector,x_train,x_test,y_train,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(class_weight = 'balanced')
model_evaluation(dtc,x_train,x_test,y_train,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight = 'balanced', max_depth = 15)
model_evaluation(rfc,x_train,x_test,y_train,y_test)
```
