```python
df.drop(columns = ['ID', 'Unnamed: 0'], inplace = True)
```

```python
X = df.drop(columns = ['class'])
y = df['class']
```

```python
X
```

```python
X_cleaned = X.copy()
```

```python
# Impute missing values: median if outliers exist, mean otherwise
for col in X.columns:
    series = X_cleaned[col]
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    has_outliers = ((series < lower_bound) | (series > upper_bound)).any()

    if has_outliers:
        impute_value = series.median()
    else:
        impute_value = series.mean()

    X_cleaned[col].fillna(impute_value, inplace=True)

# Confirm all missing values are handled
missing_after_imputation = X_cleaned[X.columns].isnull().sum().sum()
missing_after_imputation
```

```python
X_cleaned.describe()
```

```python
# Check class distribution
class_distribution = df['class'].value_counts()
class_distribution_percentage = df['class'].value_counts(normalize=True) * 100

# Combine and display
class_distribution_df = pd.DataFrame({
    'Count': class_distribution,
    'Percentage': class_distribution_percentage.round(2)
})
print(class_distribution_df)
```

```python
from sklearn.preprocessing import RobustScaler

# Columns to scale (NDVI features)
ndvi_columns = X_cleaned.columns  # assuming first 3 columns are ID, Unnamed, class

# Initialize and apply RobustScaler
scaler = RobustScaler()
X_cleaned[ndvi_columns] = scaler.fit_transform(X_cleaned[ndvi_columns])

# Optional: View summary of scaled features
print(X_cleaned[ndvi_columns].describe())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
# Step 1: Encode target class
le = LabelEncoder()
y = le.fit_transform(y)

# Step 2: Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_cleaned, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
corr = X_train.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
X_train_reduced = X_train.drop(columns=to_drop)
X_test_reduced = X_test.drop(columns=to_drop)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
# Keep interaction_only=True to avoid unnecessary explosion
poly = PolynomialFeatures(degree=3, interaction_only=True, include_bias=False)
X_train_poly = poly.fit_transform(X_train_reduced)
X_test_poly = poly.transform(X_test_reduced)

# Fit your best logistic model here
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    C=100,
    class_weight='balanced',
    max_iter=1000
)
model.fit(X_train_poly, y_train)
y_pred = model.predict(X_test_poly)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
